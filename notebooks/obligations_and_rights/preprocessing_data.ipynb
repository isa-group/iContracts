# Data preprocessing notebook


```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
```

```python
file_name = "test.xlsx"
csv_name = "test.csv"
```

```python
final_data_raw=pd.read_excel(file_name)
```

```python
final_data_raw
```

```python
df = pd.DataFrame(final_data_raw)
df.head()
```

```python
df['text'] = df['text'].str.replace(r'\s+', ' ', regex=True)  # Eliminar espacios adicionales
df['text'] = df['text'].str.strip()  # Eliminar espacios al principio y al final
df['text'] = df['text'].str.replace(r'\.\s+', '. ', regex=True)  # Asegurar espacios después de los puntos
```

```python
df.head()
```

```python
df.drop_duplicates(subset='Segment', inplace=True)
df
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

df['Tokens'] = df['Segment'].str.split()  # Tokenización
df['Tokens'] = df['Tokens'].apply(lambda x: [word for word in x if word not in stop_words])  # Eliminar palabras vacías
```

```python
df['Clean_sentences'] = df['Tokens'].apply(lambda x: ' '.join([str(token) for token in x]))
```
